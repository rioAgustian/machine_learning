## Anomaly Detection

### Rio Agustian Gilang Fernando

\begin{align}
p(x) = p(x_1, \mu_1, \sigma_1^2)* p(x_2, \mu_2, \sigma_2^2)... *p(x_n, \mu_n, \sigma_n^2)
\end{align}

Kalau $p(x) < \varepsilon$, maka termasuk anomali

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.random.normal(loc=5, scale=1, size=50)
x2 = np.random.normal(loc=5, scale=1, size=50)

plt.scatter(x1, x2, marker="+");
```

```python
from mpl_toolkits import mplot3d

def g(x, mu, sigma):
    part1 = np.sqrt(2 * np.pi * sigma**2)
    part2 = np.exp(-(x-mu)**2 / (2 * sigma**2))
    return 1/part1 * part2

x = np.arange(0, 10, 0.01)
y = np.arange(0, 10, 0.01)
X, Y = np.meshgrid(x, y)

# Plot 3d probabilitas munculnya nilai x1 dan x2
p = g(X, np.mean(x1), np.var(x1)) * g(Y, np.mean(x2), np.var(x2))

ax = plt.axes(projection ='3d')
ax.plot_surface(X, Y, p, cmap='jet')
plt.show()
```
