## **Regresi Linear dengan *Gradient Descent***

#### **Rio Agustian Gilang Fernando**

Persamaan linear

\begin{align}
f_{w,b}(x^{(i)}) = wx^{(i)} + b
\end{align}

Menentukan nilai $w$ dan $b$ dengan menggunakan *gradient descent*
\begin{align}
w &= w - \alpha \frac{1}{m}\sum_{i=1}^n (f_{w,b}(x^{(i)})-y^{(i)}) x^{(i)} \\  
b &= b - \alpha \frac{1}{m}\sum_{i=1}^n (f_{w,b}(x^{(i)})-y^{(i)})   
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.05)
y = 1*x + 1 + np.random.normal(0, 3, size=x.shape)

def f(x, w, b):
    return w*x + b

w, b = 0, 0
alpha, tol, eps = 0.001, 0.001, 1
while eps > tol:
    tmp_w = w - alpha/x.shape[0] * np.sum((f(x, w, b) - y)*x)
    tmp_b = b - alpha/x.shape[0] * np.sum((f(x, w, b) - y))
    eps   = (tmp_w - w)/tmp_w
    w = tmp_w
    b = tmp_b

plt.scatter(x, y, color='red')
plt.plot(x, f(x, w, b));
```
