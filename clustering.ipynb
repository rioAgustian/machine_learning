Algoritma clustering sederhana. Hanya ada 2 cluster.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.arange(0, 7, 0.1)
x2 = np.arange(8, 15, 0.1)z

y1 = 1 * x1 + np.random.normal(loc=0, scale=3, size=x1.shape[0])
y2 = 1 * x2 + np.random.normal(loc=0, scale=3, size=x2.shape[0])

x3 = np.concatenate((x1, x2))
y3 = np.concatenate((y1, y2))

plt.scatter(x3, y3);
```

```python
# Assign initial random cluster centroids
mu0 = [np.random.uniform(0, 15), np.random.uniform(0, 15)]
mu1 = [np.random.uniform(0, 15), np.random.uniform(0, 15)]

for i in range(10):
    # Assign points to cluster centroids
    red  = []
    blue = []
    for i in range(x3.shape[0]):
        # This part calculate distance
        distance_0 = np.sqrt((x3[i]-mu0[0])**2 + (y3[i]-mu0[1])**2)
        distance_1 = np.sqrt((x3[i]-mu1[0])**2 + (y3[i]-mu1[1])**2)
        if distance_0 < distance_1:
            red.append([x3[i], y3[i]])
        else:
            blue.append([x3[i], y3[i]])

    # Move the cluster centroids
    mu0[0] = np.mean([row[0] for row in red])
    mu0[1] = np.mean([row[1] for row in red])
    
    mu1[0] = np.mean([row[0] for row in blue])
    mu1[1] = np.mean([row[1] for row in blue])
    
red  = np.array(red)
blue = np.array(blue)

plt.scatter(red[:, 0], red[:, 1], color="red")
plt.scatter(blue[:, 0], blue[:, 1], color="blue")
plt.xlim(0, 15)
plt.ylim(0, 15)
plt.show()
```
